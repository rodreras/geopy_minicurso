## Matplotlib e Seaborn

Matplotlib é uma biblioteca criada para fazer visualizações estáticas, animadas, interativas com Python. [O Guia do Usuário pode ser encontrado aqui](https://matplotlib.org/stable/users/index).

Já o [Seaborn](https://seaborn.pydata.org/), é uma outra biblioteca, baseada na Matplotlib. Proporciona uma interface de alto nível para criar gráficos informativos e estatísticos.

```python
#vamos importar um módulo específico da biblioteca matplotlib
import matplotlib.pyplot as plt

#importando seaborn
import seaborn as sb #sns

#importando numpy para criar os dados
import numpy as np

#importando os para lidar com as pastas na hora de salvar
import os

#função mágica para fazer aparecer os gráficos no notebook
%matplotlib inline
```

Para os exemplos de visualização, usaremos as principais funções:

- `subplots`: cria um espaço para uma ou várias figuras em linhas e/ou colunas.
- `scatter`: cria um gráfico de dispersão de pontos.
- `plot`: cria um gráfico de linha simples.
- `bar`: cria um gráfico de barras.
- `hist`: cria um histograma.
- `boxplot` e `violinplot`: criam figuras que permitem a entender a distribuição dos dados.

Com objetivo de indentificar as figuras, serão utilizados os seguintes nomes: 

- `fig`: referência ao quadro geral da figura.
- `ax`ou `axs`: referência às subfiguras criadas dentro de um quadro geral.

## 1. Diagrama de dispersão

```python
#criando dados
x = np.random.randn(1,80)
y = np.random.randn(1,80)

#figura principal
fig, ax = plt.subplots(figsize = (8,8))

#diagrama de dispersão
ax.scatter(x,y);

```

## 2. Grafico de Linhas

```python
# criando dados
x = np.arange(1, 25, 1) #start stop step
y = x ** 2
z = x ** 3


# Figura principal
fig, ax = plt.subplots(figsize=(6, 6))

# Gráfico de linhas
ax.plot(x, y);
ax.plot(x, z);
```

Para adicionar legendas e títulos, usaremos as funções a seguir:

```python
# criando dados
x = np.arange(1, 25, 1) #start stop step
y = x ** 2
z = x ** 3


# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico de linhas
ax.plot(x, y, label ="$x^{2}$" )
ax.plot(x, z, label = "$x^{3}$")

#adicionando legenda
ax.legend();

#adicionando título do gráfico
ax.set_title('Gráfico exponencial', fontsize = 14);

#titulo superior - aplicado para o quadro geral
fig.suptitle('Título Superior', fontsize = 20);

#adicionando o nome do eixo x e y
plt.xlabel('Eixo X');
plt.ylabel('Eixo Y');

## adicionando grades no gráfico
plt.grid(alpha = 0.5); #alpha é para deixar um pouco transparente
```

## 5. Boxplot

```python
# Input
x = np.random.randn(10000,)
y = 2*x + 1
z = 2 - x


# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Boxplot
ax.boxplot([x, y, z]);

plt.grid(alpha = 0.3);
```

## 6. Violinplot

```python
# Input
x = np.random.randn(10000,)
y = 2*x + 1
z = 2 - x


# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Violinplot
ax.violinplot([x, y, z]);
```

## 7. Modificando seu gráfico

Em algumas situações, existe a necessidade de realizar modificações e estilizações em nossos gráficos.

É claro que podemos fazer isso. Alterar cores, tamanho dos símbolos, transparência, entre muitas outras coisas.

Aqui, vamos passar apenas por alguns exemplos, mas caso queira saber mais, veja a documentação tanto do Matplotlib quanto do Seaborn para entender melhor como funciona.

### 7.1 Alterando o tamanho dos pontos

Para modificar o tamanho de um ponto num gráfico de dispersão, utilizamos o parâmetro `size`, que também pode ser abreviado para `s`. Particularmente, prefiro usar o nome da função completo, mas isso pode mudar de acordo com sua preferência.

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randint(5, 100, (1000))

# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico de dispersão
im = ax.scatter(x,
                y,
                s = z
                );
```

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randint(5, 100, (1000))

# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico de dispersão
  #adicionalmente, vamos alterar outras coisas, como borda e transparência
im = ax.scatter(x,
                y,
                s = z,
                alpha = 0.75,
                edgecolor = 'black'
                );
```

```python
# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico
im = ax.scatter(x, y, s=z, edgecolor="black", alpha=0.8)

#Definindo a Legenda
handles, labels = im.legend_elements(prop="sizes")

# Escrevendo a legenda
ax.legend(title = 'Group Size',
          handles = handles, 
          labels = labels);
```

### 7.2 Barra de Cores

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randn(1000)
size = np.random.randint(5, 100, (1000))

# Mapa de cores
cmap = plt.cm.inferno


# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico
im = ax.scatter(x,
                y,
                c=z,
                cmap=cmap,
                s=50,
                edgecolor="black");

#Definindo a Legenda
handles, labels = im.legend_elements(prop="colors")

# Escrevendo a legenda
ax.legend(title = 'Group Size',
          handles = handles, 
          labels = labels);
```

```python
# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# Gráfico
im = ax.scatter(x, y, c=z, cmap=cmap, s=50, edgecolor="black")

# Barra de cores
cbar = fig.colorbar(im, ax=ax)
cbar.ax.set_title("Cu(%)", fontsize=18, y=1.01)
```
