```python
#Carregue o arquivo sedimentos_patos.csv

sedPato = pd.read_csv(caminho+'/'+'sedimentos_pato.csv')
sedPato.head()
```

```python
#Crie uma cópia
patoCopy = sedPato.copy()

patoCopy.info()
```

```python
#Crie uma variável chamada ag e em seguida filtre apenas os dados de prata
ag = patoCopy.filter(items = ['Estação',
                              'N__Lab_ ',
                              'X',
                              'Y',
                              'Folha',
                              'Ag (ppm)'])
ag
```

```python
#Crie um histograma 

# Figura principal
fig, ax = plt.subplots(figsize=(8,8))


x = ag['Ag (ppm)']

# Histograma
ax.hist(x,
        bins=15,
        color = 'red');

ax.set_title('Histograma para os dados de Prata (ppm)',
             fontsize = 14);

plt.ylabel('Número de ocorrências');
plt.grid(axis = 'y', alpha = 0.3);
```

```python
#Crie um boxplot

#criando a figura
fig, ax = plt.subplots(figsize=(8,8))

ax.boxplot(x);
plt.grid(alpha = 0.5);
ax.set_title('Box Plot para dados de Prata (ppm)');
plt.xlabel('Prata');
plt.ylabel('Concentração em ppm');
```

```python
#Crie uma variável chamada relation e filtre valores de Ag e Cu
relation = patoCopy.filter(items = ['Estação',
                              'N__Lab_ ',
                              'X',
                              'Y',
                              'Folha',
                              'Ag (ppm)',
                              'Cu (ppm)'])
relation.head()
```

```python
#Crie um quadro geral com 4 gráficos

fig, ax = plt.subplots(2,2, figsize = (8,8))

#Nos gráficos superiores, coloque o histograma de cada elemento. 
#Nos dois inferiores, os box-plots.

#grafico superior direito
ax[0,1].hist(relation['Cu (ppm)']);
ax[0,1].set_title('Histograma para Cu (ppm)');

#grafico inferior esquerdo
ax[1,1].boxplot(relation['Cu (ppm)']);
ax[1,1].set_title('Box-plot para Cu (ppm)');

#grafico superior esquerdo
ax[0,0].hist(relation['Ag (ppm)']);
ax[0,0].set_title('Histograma para Ag (ppm)');

#grafico inferior esquerdo
ax[1,0].boxplot(relation['Ag (ppm)']);
ax[1,0].set_title('Box-plot para Ag (ppm)');
```

```python
 #Crie um gráfico de dispersão entre Cu e Au
fig, ax = plt.subplots(figsize = (8,8))

#definindo a cor
cmap = plt.cm.hsv

#criando o grafico
plot = ax.scatter(x = relation['Ag (ppm)'],
           y = relation['Cu (ppm)'],
           cmap = cmap,
           edgecolor = 'black',
           s = 50
           );

#colocando nome nos eixos
plt.xlabel('Concentração Ag (ppm)');
plt.ylabel('Concentração Cu (ppm)');

#titulo do grafico
ax.set_title('Gráfico de Dispersão Ag x Cu');

#grid
plt.grid(alpha = 0.3);

```
